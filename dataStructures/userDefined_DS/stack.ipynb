# Stack
stack is a ordered collection of items, think of it as a pile of dishes...

```python
class Stack:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return self.items == []
    
    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("Stack is empty")
    
    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            raise IndexError("Stack is empty")
    
    def size(self):
        return len(self.items)
```

```python
stack1 = Stack()
stack1.size()
```

```python
stack1.push(10)
print("size of the stack is", stack1.size())
print("top element of the stack is", stack1.peek())
stack1.push(19)
print("top element of the stack is", stack1.peek())
stack1.pop()
stack1.pop()
```
